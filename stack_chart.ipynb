```python
df = pd.read_excel('stack_chart.xlsx')
df.head()
```

```python
cate_price_pivot = pd.pivot_table(data = df,
                index = 'cate_1', columns = 'cate_2', 
                values = 'amt').fillna(0)
cate_price_pivot
```

```python
_colors = random.sample(color_lst, len(cate_price_pivot.columns))
```

```python
ax = cate_price_pivot.plot.barh(stacked=True, figsize=(10,7),
                            color = _colors)
ax.xaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
plt.xticks(rotation = 45)
plt.title('제목을 입력하시오')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
