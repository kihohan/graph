```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import matplotlib.ticker as tkr
import matplotlib as mpl

import warnings
warnings.filterwarnings(action='ignore')
```

```python
df = pd.read_excel('double_y.xlsx')
```

```python
df.head()
```

```python
plt.figure(figsize = (30, 10))
ax1 = sns.pointplot(data = df, x = 'date', y = 'cnt')
ax1.set_xticklabels(df['date'], rotation=90, ha='right')
ax1.legend(['y1'],bbox_to_anchor=(1,1.1))

ax2 = ax1.twinx()
ax2 = sns.pointplot(data = df, x = 'date', y = 'amt', color = 'yellow')
ax2.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
ax2.legend(['y2'],bbox_to_anchor=(1,1.15))

ax2.axvspan(0, 30, alpha=0.5, color='bisque')
ax2.axvspan(30, 58, alpha=0.5, color='lightpink')
ax2.axvspan(58, 89, alpha=0.5, color='bisque')
ax2.axvspan(89, 119, alpha=0.5, color='lightpink')
ax2.axvspan(119, 150, alpha=0.5, color='bisque')

plt.title('double_y\n', fontsize = 20)
plt.tight_layout()
plt.show()
```
